NumPy Array Inspection and Creation - Programming Questions

Array Inspection:

1.	Write a Python program to create a NumPy array from a list and print its shape.
o	List = [1, 2, 3, 4, 5, 6]


```python
import numpy as np
List = [1, 2, 3, 4, 5, 6]

a = np.array(List)
print(a)
print(a.shape)
```

2.	Write a Python program to create a NumPy array and print its number of dimensions (ndim).
o	Array = [[1,2,3],[4,5,6]]


```python
import numpy as np

Array = [[1,2,3],[4,5,6]]
b = np.array(Array)
print(b)
print(b.ndim)
```

3.	Write a Python program to create a NumPy array and print its total number of elements (size).
o	Array = [10, 20, 30, 40, 50]


```python
import numpy as np
Array = [10, 20, 30, 40, 50]

c = np.array(Array)
print(c)
print(c.size)
```

4.	Write a Python program to find the data type (dtype) of a given NumPy array.
o	Array = [1.0, 2.5, 3.8]


```python
import numpy as np

Array = [1.0, 2.5, 3.8]

d = np.array(Array)
print(d.dtype)
```

5.	Write a Python program to convert the data type of a NumPy array from float to int using astype().
o	Array = [5.5, 6.7, 8.9]


```python
import numpy as np

Array = [5.5, 6.7, 8.9]
e = np.array(Array,dtype = int)
e.astype(int)
print(e)
print(e.dtype)
```
