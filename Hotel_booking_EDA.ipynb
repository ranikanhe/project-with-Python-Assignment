### Chart - 2
####*Question no. 2 Which year had highest bookings[ City hotels or Resort Hotels ]*

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (11, 9)
matplotlib.rcParams['figure.facecolor'] = '#F5F5F5'

sns.countplot(x='arrival_date_year', data=df1, hue='hotel')
plt.title('Yearly Bookings')
plt.show()
```

This chart illustrates the yearly hotel bookings for **Resort Hotels** and **City Hotels** from **2015 to 2017**. The x-axis represents the arrival year, while the y-axis shows the total number of bookings. In 2015, both hotel types received almost an equal number of bookings, indicating similar demand. In 2016, bookings increased sharply for both hotels, with City Hotels recording the highest number of bookings overall, showing a strong customer preference. In 2017, bookings declined compared to 2016, but City Hotels continued to have more bookings than Resort Hotels. Overall, the chart indicates that **City Hotels were consistently more popular than Resort Hotels**, and **2016 was the peak year for hotel bookings**.



### Chart - 3
####*Question no. 3 Which Type Of hotel have Highest revenue or ADR[ City hotels or Resort Hotels ]*

```python
# Chart - 3 visualization code (Highest revenue)
# group by hotel
grup_by_hotel=df1.groupby('hotel')

#grouping by hotel adr
highest_adr=grup_by_hotel['adr'].mean().reset_index()

#set plot size
plt.figure(figsize=(11,9))

#plot the graph
sns.barplot(x=highest_adr['hotel'],y=highest_adr['adr'])

# set labels
plt.xlabel('Hotel type')
plt.ylabel('ADR')
plt.title("Avg ADR of each Hotel type")
```

This chart shows the **average ADR (Average Daily Rate)** for each hotel type, comparing **City Hotels** and **Resort Hotels**. The x-axis represents the hotel type, while the y-axis shows the ADR value. From the chart, it is clear that **City Hotels have a higher average ADR** than Resort Hotels, indicating that the average room price per night is higher in City Hotels. Resort Hotels have a comparatively lower ADR, suggesting more affordable average pricing. Overall, the chart highlights that **City Hotels generate higher room revenue per booking compared to Resort Hotels**.


### Chart - 4
####*Question no. 4 What is the percentage distribution of required car parking space*

```python
# Chart - 4 visualization code (required car parking space)


import matplotlib.pyplot as plt

# Count values
parking_counts = df1['required_car_parking_spaces'].value_counts()

# Explode all slices slightly
explode = [0.05] * len(parking_counts)

# Plot pie chart
plt.figure(figsize=(11, 9))
parking_counts.plot(
    kind='pie',
    explode=explode,
    autopct='%1.1f%%',
    shadow=False,
    startangle=90,
    fontsize=15
)

# Title and legend
plt.title('% Distribution of Required Car Parking Spaces', fontsize=18)
plt.ylabel('')  # Remove default y-label
plt.legend(
    labels=parking_counts.index,
    bbox_to_anchor=(0.9, 1),
    loc='upper left'
)

plt.show()
```

The image shows a pie chart titled **“% Distribution of Required Car Parking Spaces.”** It illustrates how frequently different numbers of required car parking spaces occur in the dataset. The chart is dominated by the category **0 parking spaces**, which accounts for **91.6%** of the total, indicating that the vast majority of hotels do not require any car parking space. The next most common category is **1 parking space**, representing **8.3%**, shown as a small but noticeable slice. The remaining categories (**2, 3, and 8 parking spaces**) contribute an almost negligible proportion, collectively close to **0%**, and appear as very thin slices. Overall, the visualization clearly highlights that zero required parking spaces overwhelmingly prevail compared to other categories.

### Chart - 7
####*Question no. 7 -In which months there are more number of bookings*

```python
# Chart - 7 visualization code
# groupby arrival_date_month and taking the hotel count
bookings_by_months_df=df1.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts"})
# Create list of months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# creating df which will map the order of above months list without changing its values.
bookings_by_months_df['arrival_date_month']=pd.Categorical(bookings_by_months_df['arrival_date_month'],categories=months,ordered=True)
# sorting by arrival_date_month
bookings_by_months_df=bookings_by_months_df.sort_values('arrival_date_month')

bookings_by_months_df
```

```python
# set plot size
plt.figure(figsize=(11,9))

#pltting lineplot on x- months & y- booking counts
sns.lineplot(x=bookings_by_months_df['arrival_date_month'],y=bookings_by_months_df['Counts'])

# set title for the plot
plt.title('Number of bookings across each month')
#set x label
plt.xlabel('Month')
#set y label
plt.ylabel('Number of bookings')
```

```python
df1['arrival_date_month'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(11,9),fontsize=20)
```

The image displays a pie chart representing the **monthly distribution of hotel bookings** across the year. Each slice corresponds to a month, with percentages indicating the share of total bookings. **August** has the highest proportion at **12.9%**, followed by **July (11.5%)**, showing peak booking activity during the mid-year period. Moderate booking levels are observed in **May (9.6%)**, **April (9.1%)**, and **June (8.9%)**. The remaining months—**March (8.6%)**, **October (7.9%)**, **September (7.7%)**, **February (7.0%)**, **December (5.9%)**, **November (5.7%)**, and **January (5.4%)**—show relatively lower and more evenly distributed booking shares. Overall, the chart highlights a seasonal trend, with higher hotel demand during summer months and lower activity at the beginning and end of the year.
