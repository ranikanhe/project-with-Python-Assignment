```python
import pandas as pd

# Load your file
df = pd.read_csv("/content/drive/MyDrive/SQL/sales_data.csv")

# 31. Remove the EmployeeID column
df = df.drop(columns=["EmployeeID"])

# 32. Reset the index of the DataFrame
df = df.reset_index(drop=True)

# 33. Replace employees who did not achieve the target
df["TargetAchieved"] = df["TargetAchieved"].replace({"No": "Needs Improvement"})

# 34. Create PerformanceLevel column
df["PerformanceLevel"] = df["SalesAmount"].apply(
    lambda x: "High" if x > 40000 else ("Medium" if x >= 20000 else "Low")
)

# 35. Convert Age column to integer
df["Age"] = df["Age"].astype(int)

# 36. Create AboveAverageEarnings column
average_earnings = df["TotalEarnings"].mean()
df["AboveAverageEarnings"] = df["TotalEarnings"].apply(
    lambda x: "Yes" if x > average_earnings else "No"
)

# 37. Convert Department column to categorical
df["Department"] = df["Department"].astype("category")

# 38. Top 5 employees based on Total Earnings
df.nlargest(5, "TotalEarnings")


# 39. Bottom 5 employees based on Total Earnings
df.nsmallest(5, "TotalEarnings")


# 40. Save the modified DataFrame to a new CSV file
df.to_csv("/content/drive/MyDrive/SQL/sales_data_modified.csv", index=False)

print("\nFile saved successfully as: sales_data_modified.csv")
```
