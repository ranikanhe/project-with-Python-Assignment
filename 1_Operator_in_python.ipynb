```python
# 15. Write a Python program to calculate the simple interest using the formula: SI = (P ×
# R × T) / 100, where P is the principal amount, R is the rate of interest, and T is the
# time in years.

# Take inputs from the user
P = float(input("Enter Principal amount: "))
R = float(input("Enter Rate of interest: "))
T = float(input("Enter Time in years: "))

# Calculate Simple Interest
SI = (P * R * T) / 100

# Display result
print("Simple Interest is:", SI)
```
