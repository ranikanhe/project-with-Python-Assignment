```python
#1.Given a list [4, 8, 2, 10, 6], find the maximum number using a for loop.
a = [4, 8, 2, 10, 6]
m = a[0]

for i in a:
    if i > m:
        m = i

print("Maximum number =", m)
```

```python
#2.Take input from the user and find the factorial of the number using a for loop.
n=int(input("Enter a number:"))
fact=1

for i in range(1,n+1):
  fact*=1
print("Factorial=",fact)
```

```python
#3.Given a list ['apple', 'banana', 'cherry', 'date', 'fig'], print each element along with its index using a for loop.
fruits = ['apple', 'banana', 'cherry', 'date', 'fig']

for i in range(len(fruits)):
    print(i, fruits[i])
```

```python
#4.Take a string input from the user and count the number of uppercase letters using a for loop. (Use char.isupper()) ðŸ˜Š
s = input("Enter a string: ")
count = 0

for ch in s:
    if ch.isupper():
        count += 1

print("Uppercase letters =", count)
```

```python
#5.Given a list of integers [17, 25, 39, 42, 58, 71, 86, 93], find the sum of all the even numbers using a for loop.
a = [17, 25, 39, 42, 58, 71, 86, 93]
s = 0

for i in a:
    if i % 2 == 0:
        s += i

print("Sum of even numbers =", s)
```

```python
#6.Given a list [3, 6, 9, 12, 15], calculate the product of all the elements using a for loop.
a = [3, 6, 9, 12, 15]
p = 1

for i in a:
    p *= i

print("Product =", p)
```

```python
#7.Given a list [2, 3, 5, 7, 11, 13, 17, 19], find the index of the first occurrence of the number 7 using a for loop.
a = [2, 3, 5, 7, 11, 13, 17, 19]

for i in range(len(a)):
    if a[i] == 7:
        print("Index of 7 =", i)
        break
```

```python
#8.Take a string input from the user and reverse it using a for loop.
s = input("Enter a string: ")
rev = ""

for ch in s:
    rev = ch + rev

print("Reversed string =", rev)
```

```python
#9.Given a list [10, 20, 30, 40, 50], print each element multiplied by 2 using a for loop.
a = [10, 20, 30, 40, 50]

for i in a:
    print(i * 2)
```

```python
#10.Given a list of words ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'], find the longest word using a for loop.
words = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
longest = words[0]

for w in words:
    if len(w) > len(longest):
        longest = w

print("Longest word =", longest)
```

```python
#11.Take a string input from the user and count the number of vowels using a for loop. (Vowels: a, e, i, o, u)
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("Number of vowels =", count)
```

```python
#12.Given a list of integers [4, 9, 16, 25, 36], print the square root of each number using a for loop.
import math
a = [4, 9, 16, 25, 36]

for i in a:
    print(math.sqrt(i))
```

```python
#13.Given a list [1, 3, 5, 7, 9, 11], check if a user-inputted number is present in the list using a for loop.
a = [1, 3, 5, 7, 9, 11]
n = int(input("Enter a number: "))
found = False

for i in a:
    if i == n:
        found = True
        break

if found:
    print("Number is present")
else:
    print("Number is not present")
```

```python
#15.Take a sentence as input from the user and count the number of words using a for loop. (Words are separated by spaces.)
s = input("Enter a sentence: ")
count = 1

for ch in s:
    if ch == " ":
        count += 1

print("Number of words =", count)
```

```python
#16.Given a list [20, 40, 60, 80, 100], create a new list with elements divided by 10 using a for loop.
a = [20, 40, 60, 80, 100]
new_list = []

for i in a:
    new_list.append(i / 10)

print(new_list)
```

```python
#17.Given a list ['red', 'green', 'blue', 'yellow', 'purple'], print only the words that have more than 4 letters using a for loop.
colors = ['red', 'green', 'blue', 'yellow', 'purple']
for c in colors:
    if len(c) > 4:
        print(c)
```

```python
#18.Take a number as input from the user and generate a list of its multiplication table up to 10 using a for loop.
n = int(input("Enter a number: "))
table = []

for i in range(1, 11):
    table.append(n * i)

print(table)
```

```python
#19.Given a dictionary {'A': 10, 'B': 20, 'C': 30, 'D': 40}, print all key-value pairs using a for loop.
d = {'A': 10, 'B': 20, 'C': 30, 'D': 40}

for key in d:
    print(key, "=", d[key])
```
