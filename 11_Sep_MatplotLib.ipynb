```python
# 1.	Create a bar chart showing the number of employees in each department.

emp_count = df['Department'].value_counts()

plt.figure(figsize = (7,3))

e = emp_count.plot(kind = 'bar',color = ['yellow','red'])
e.bar_label(e.containers[0], label_type='edge')
plt.title('Number Of Employees per Departments',color = 'red')
plt.xlabel('Departments')
plt.ylabel('Counts')
plt.xticks(rotation = 30)
plt.show()
```

```python
#11. Generate a violin plot to visualize salary distribution by gender.

# Separate salary data by gender
male_salary = df[df["Gender"] == "Male"]["Salary"]
female_salary = df[df["Gender"] == "Female"]["Salary"]

# Create violin plot
plt.figure()
plt.violinplot([male_salary, female_salary], showmeans=True)

plt.xticks([1, 2], ["Male", "Female"])
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.title("Salary Distribution by Gender")

plt.tight_layout()
plt.show()
```

```python
#12.Plot a grouped bar chart comparing average salary per department.


# Calculate average salary per department
avg_salary = df.groupby("Department")["Salary"].mean()

# Plot grouped bar chart
plt.figure()
avg_salary.plot(kind="bar")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Average Salary per Department")
plt.show()
```

```python
#13.Create a bubble chart where the bubble size represents the bonus amount and position shows salary.

# Create bubble chart
plt.figure()
plt.scatter(
    x=df["Department"],
    y=df["Salary"],
    s=df["Bonus"],   # bubble size represents bonus
    alpha=0.6
)

plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Bubble Chart: Salary vs Department (Bubble Size = Bonus)")
plt.show()
```

```python
#15.Create a dual-axis chart where one axis shows salary distribution and the other shows employee count.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Python/employee_data.csv")

# Calculate average salary per department
avg_salary = df.groupby("Department")["Salary"].mean()

# Calculate employee count per department
emp_count = df.groupby("Department").size()

# Create figure and first axis
fig, ax1 = plt.subplots()

# Plot average salary (left Y-axis)
ax1.bar(avg_salary.index, avg_salary.values)
ax1.set_xlabel("Department")
ax1.set_ylabel("Average Salary")

# Create second Y-axis
ax2 = ax1.twinx()

# Plot employee count (right Y-axis)
ax2.plot(emp_count.index, emp_count.values, marker='o')
ax2.set_ylabel("Employee Count")

# Title
plt.title("Dual-Axis Chart: Salary Distribution and Employee Count")

plt.show()
```

```python
#16.Create a KDE (Kernel Density Estimate) plot for salary distribution.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load employee data
df = pd.read_csv("/content/drive/MyDrive/Python/employee_data.csv")

# Create KDE plot for Salary
plt.figure()
sns.kdeplot(df["Salary"], fill=True)
plt.xlabel("Salary")
plt.ylabel("Density")
plt.title("KDE Plot for Salary Distribution")
plt.show()
```

```python
#17.Generate a 3D scatter plot to visualize salary, bonus, and years of experience.
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assumed employee experience data
data = {
    "Experience": [1, 3, 5, 7, 10],
    "Salary": [30000, 45000, 60000, 75000, 90000],
    "Bonus": [2000, 4000, 6000, 8000, 10000]
}

df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["Experience"], df["Salary"], df["Bonus"])

ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salary")
ax.set_zlabel("Bonus")
ax.set_title("3D Scatter Plot: Experience vs Salary vs Bonus")

plt.show()
```

```python
#19.Plot a grouped violin plot showing salary distribution across departments.

import pandas as pd
import matplotlib.pyplot as plt

# STEP 1: Load employee data
df = pd.read_csv("/content/drive/MyDrive/Python/employee_data.csv")

# STEP 2: Department-wise salary data
departments = df["Department"].unique()
salary_data = [df[df["Department"] == d]["Salary"] for d in departments]

# STEP 3: Grouped violin plot
plt.figure(figsize=(8, 4))
plt.violinplot(salary_data, showmeans=True)

plt.xticks(range(1, len(departments)+1), departments)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Salary Distribution Across Departments")

plt.tight_layout()
plt.show()

```
